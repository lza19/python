```python
class user:
    def __init__(self,name, nam):
        self.age = name
        self.country = nam

    def show(self):
        print("Age :",self.age, " Country :",self.country)

    def __del__(self):
        print("Destructor called, user object is being deleted.")

```

```python
x1 = user("veg1","USA")
x2 = user("veg","USA1")
x1.show()
x2.show()
```
