```python
class Score:
    def __init__(self, name ,points):
        self.name = name
        self.points = points
    def display(self):
        print("Name :",self.name,"score :",self.points)

class scorecard:
    def __init__(self, scores):
        self.scores = scores
    def display_all(self):
        print("Scorecard :",self.scores)
```

```python
scores = {'Math': 90, 'Science': 85, 'English': 95}
```

```python
q1 = scorecard(scores)
q1.display_all()
```

```python

x1 = Score(scores,90)
x1.display()
```
